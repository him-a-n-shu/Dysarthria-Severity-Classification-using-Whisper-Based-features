```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import layers, models, optimizers
import tensorflow as tf

# =================== CONFIGURATION ====================
# Data directory path with subdirectories for each class label
DATA_DIR = '/kaggle/input/class-data/class-data'  # Modify this to your data directory
CLASSES = ['VERY LOW', 'LOW', 'MEDIUM', 'HIGH']

# Audio parameters
SR = 16000                # Sampling rate
FRAME_LENGTH_MS = 25      # Frame length in ms for MFCC extraction
HOP_LENGTH_MS = 10        # Hop length in ms for MFCC extraction
n_mfcc = 20               # Number of MFCC features

# Convert ms to samples
frame_length = int(SR * (FRAME_LENGTH_MS / 1000.0))
hop_length = int(SR * (HOP_LENGTH_MS / 1000.0))
# =======================================================

def load_data(data_dir, classes, sr=SR):
    """
    Loads .wav files from each subdirectory (each representing a class)
    and returns lists for signals and corresponding labels.
    Displays a progress bar for each class.
    """
    signals = []
    labels = []
    
    for label in classes:
        class_dir = os.path.join(data_dir, label)
        if not os.path.isdir(class_dir):
            print(f"Warning: Directory {class_dir} not found.")
            continue
        
        files = [f for f in os.listdir(class_dir) if f.lower().endswith('.wav')]
        print(f"Loading '{label}' with {len(files)} files:")
        for file in tqdm(files, desc=f"Processing {label}", unit="file"):
            file_path = os.path.join(class_dir, file)
            try:
                signal, _ = librosa.load(file_path, sr=sr)
                signals.append(signal)
                labels.append(label)
            except Exception as e:
                print(f"Error loading {file_path}: {e}")
    
    return signals, labels

def extract_mfcc_features(signals, sr=SR, n_mfcc=n_mfcc, frame_length=frame_length, hop_length=hop_length):
    """
    Extract MFCC features from a list of audio signals.
    Returns a numpy array of MFCC feature arrays.
    Each sample is padded or truncated to a fixed number of frames.
    """
    features = []
    for signal in signals:
        mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc, n_fft=frame_length, hop_length=hop_length)
        # For uniform shape: take only a fixed number of frames (e.g., first 100 frames, pad if needed).
        if mfcc.shape[1] < 100:
            pad_width = 100 - mfcc.shape[1]
            mfcc = np.pad(mfcc, ((0, 0), (0, pad_width)), mode='constant')
        else:
            mfcc = mfcc[:, :100]
        features.append(mfcc)
    return np.array(features)

def build_cnn_model(input_shape, num_classes):
    """
    Builds a CNN using the Functional API with an explicit Input layer.
    A Dense layer named 'features' is added before the final dropout and output layer.
    """
    inputs = layers.Input(shape=input_shape)
    
    x = layers.Conv2D(32, (3, 3), activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.MaxPooling2D((2, 2))(x)
    
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.MaxPooling2D((2, 2))(x)
    
    x = layers.Flatten()(x)
    # Penultimate Dense layer - name it so that we can easily extract it later.
    features = layers.Dense(128, activation='relu', name='features')(x)
    x = layers.Dropout(0.3)(features)
    
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = models.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=optimizers.Adam(),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

def plot_training_history(history):
    """Plots training accuracy and loss curves."""
    plt.figure(figsize=(12, 5))
    
    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_tsne(embeddings, labels):
    """Plots a t-SNE visualization of the feature embeddings."""
    tsne = TSNE(n_components=2, random_state=42)
    X_tsne = tsne.fit_transform(embeddings)
    
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='jet', alpha=0.7)
    plt.colorbar(scatter, label='Class Label')
    plt.title('t-SNE Visualization of Feature Embeddings')
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    plt.show()

# ------------------ DATA LOADING ------------------
signals, labels = load_data(DATA_DIR, CLASSES)

if len(signals) == 0:
    print("No audio files loaded. Exiting.")

# ------------------ FEATURE EXTRACTION ------------------
print("Extracting MFCC features...")
features = extract_mfcc_features(signals)
print("Feature extraction done. Feature shape:", features.shape)

# Add channel dimension for CNN: new shape (num_samples, height, width, channels)
features = np.expand_dims(features, -1)

# ------------------ LABEL ENCODING ------------------
le = LabelEncoder()
label_indices = le.fit_transform(labels)
num_classes = len(le.classes_)
labels_categorical = to_categorical(label_indices, num_classes=num_classes)

# ------------------ DATA SPLITTING ------------------
X_train, X_test, y_train, y_test = train_test_split(features, labels_categorical,
                                                    test_size=0.2,
                                                    random_state=42,
                                                    stratify=label_indices)

# ------------------ MODEL BUILDING ------------------
input_shape = X_train.shape[1:]
model = build_cnn_model(input_shape, num_classes)
model.summary()

# ------------------ MODEL TRAINING ------------------
history = model.fit(X_train, y_train, 
                    validation_split=0.2,
                    epochs=50,
                    batch_size=32,
                    verbose=1)

# ------------------ EVALUATION ------------------
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")

# ------------------ PLOTTING TRAINING CURVES ------------------
plot_training_history(history)

# ------------------ t-SNE VISUALIZATION ------------------
# Extract features from the penultimate Dense layer using the layer name 'features'
feature_extractor = models.Model(inputs=model.input,
                                 outputs=model.get_layer('features').output)
embeddings = feature_extractor.predict(X_test)

# Convert one-hot labels back to class indices for coloring in t-SNE plot
y_test_indices = np.argmax(y_test, axis=1)
plot_tsne(embeddings, y_test_indices)
```
